## 决策树API的使用

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split
```

### 载入数据

```python
wine  = load_wine()
```

```python
wine.keys()
```

```python
print(wine.DESCR)
```

```python
# 把数据转换成DataFrame
import numpy as np
import pandas as pd
```

```python
# 特征数据转换成DataFrame
wine_df = pd.DataFrame(wine.data , columns= wine.feature_names)
```

```python
wine_df.head()
```

```python
# 转换成Series
y = pd.Series(wine.target)
```

```python
y
```

```python
# 制作一个数据副本 X
X = wine_df.copy()
X
```

### 切分训练集和测试集

```python
# 数据切分
X_train , X_test , y_train, y_test = train_test_split(X, y ,
                                    test_size = 0.3 , random_state = 666)
```

### 建立决策树模型

```python
# 实例化决策树API
clf = DecisionTreeClassifier(max_depth=3)
```

```python
# 训练模型
clf.fit(X_train , y_train)
```

```python
# 用建立好的模型, 去查看在训练集上的分数
clf.score(X_train , y_train)
```

```python
# 模型在测试数据集上面的准确率
clf.score(X_test , y_test)
```

```python
# 查看模型建立后, 特征在建立模型中的重要请
clf.feature_importances_
```

```python
# 查看模型建立后, 特征在建立模型中的重要请
clf.feature_importances_
```

```python
# 查看模型建立后, 特征在建立模型中的重要请
clf.feature_importances_
```

```python
wine.feature_names
```

### 绘制学习曲线

```python
import matplotlib.pyplot as plt
plt.style.use('seaborn')
```

```python
# 建立列表接收数据
# 接收测试集数据
score_test = []
# 接收训练集数据
score_train = []
# max_depth 是 1 - 10
for i in range(1,11):
    # 建模
    clf = DecisionTreeClassifier(max_depth= i , random_state=666 )
    # 训练
    clf.fit(X_train , y_train)
    # 计算训练集分数
    score_train.append(clf.score(X_train , y_train))
    # 计算测试集分数
    score_test.append(clf.score(X_test , y_test))
```

```python
score_train
```

```python
score_test
```

```python
# 绘制训练集和测试集的分数变化曲线
plt.plot(range(1,11) , score_train , color= 'red')
plt.plot(range(1,11) , score_test , color= 'green')
```

```python
# 建立列表接收数据
# 接收测试集数据
score_test = []
# 接收训练集数据
score_train = []
# max_depth 是 1 - 10
for i in range(2,20):
    # 建模
    clf = DecisionTreeClassifier(min_samples_leaf= i , random_state=666 )
    # 训练
    clf.fit(X_train , y_train)
    # 计算训练集分数
    score_train.append(clf.score(X_train , y_train))
    # 计算测试集分数
    score_test.append(clf.score(X_test , y_test))
# 绘制训练集和测试集的分数变化曲线
plt.plot(range(2,20) , score_train , color= 'red')
plt.plot(range(2,20) , score_test , color= 'green')
```

### 使用make_blobs去创建一个不均衡数据集

```python
class_1 = 500 # 生成两类数据
class_2 = 50 # 第一类500个 ,第二类 50个
```

```python
# 设定两个类别数据的中心点
centers = [[0,0] , [2.0, 2.0]]
```

```python
clusters_std = [1.5 , 0.5] # 设定两个聚类群的标准差, 标准差越大数据越松散
```

```python
# X 就是 数据集  y 对应的分类
X , y = make_blobs(n_samples = [class_1 , class_2] , centers=centers , 
          cluster_std=clusters_std )
```

```python
X
```

```python
y
```

```python
plt.figure(figsize=(10,8))
# 把图像绘制出来
# c 如果传递一个序列, 就会按照序列的数据映射不同的颜色
# cmap  颜色映射方案
plt.scatter(X[: , 0] , X[: , 1] , c =y , cmap='rainbow')
```

```python
# 决策树建模
# 使用决策树进行建模, 找到在训练集上比较好的分数结果
clf = DecisionTreeClassifier(max_depth=4 , random_state=666 )
clf.fit(X,y)
clf.score(X,y)
```

```python
y
```

```python
# 重新建模, 使用 class_weight 这个参数 去进行建模
clf = DecisionTreeClassifier(max_depth=4 , random_state=666, 
                            class_weight={0: 1, 1: 10})
# {0: 1, 1: 10}  前边的0 和 1 分别代表他的分类标签
# 1 ,和 10 代表权重   一个0样本 权重为1  一个1样本 权重为10
clf.fit(X,y)
clf.score(X,y)
```

### 对于样本分类不平衡的情况下, 更适合于使用其他的评估指标

```python
# 导入混淆矩阵的函数
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score
```

```python
# clf_noweight 无样本权重模型
clf_noweight = DecisionTreeClassifier(max_depth=4 , random_state=666 )
clf_noweight.fit(X,y)
clf_noweight.score(X,y)
```

```python
# clf_noweight 有样本权重模型
clf_weight = DecisionTreeClassifier(max_depth=4 , random_state=666,
                                    class_weight={0: 1, 1: 10})
clf_weight.fit(X,y)
clf_weight.score(X,y)
```

```python
# 使用第一个模型对样本进行预测
y_pred = clf_noweight.predict(X)
```

```python
# 无权重的
confusion_matrix(y , y_pred )
```

```python
# 无权重的
# 因为少的类50个是标签为1 , 500个标签是0 
# 我想让50个的处于混淆矩阵的左上角, 模型labels = [0 , 1]
# 改成 labels = [1 , 0]
confusion_matrix(y , y_pred , labels= [ 1, 0 ] )
```

```python
(49 + 485)/(49 + 1 + 15 + 485 )
```

```python
49 / (49 + 1)
```

```python
# 无权重的召回率
recall_score(y , y_pred)
```

```python
# 无权重的精准度
precision_score(y , y_pred)
```

```python
# 有权重情况下的混淆矩阵
# 使用第一个模型对样本进行预测
y_pred_weight = clf_weight.predict(X)
confusion_matrix(y , y_pred_weight , labels= [ 1, 0 ] )
```

```python
#  有权重的情况下的召回率
recall_score(y , y_pred_weight)
```

```python
# 有权重情况下的精准度
precision_score(y , y_pred_weight)
```

```python
# 有权重
f1_score(y , y_pred)
```

```python
# 无权重
f1_score(y , y_pred_weight)
```

## 泰坦尼克号海难生存人员预测

### 导入所需要的库

```python
# 导入所需要的库
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import cross_val_score
import matplotlib.pyplot as plt
```

### 导入数据集，探索数据

```python
# 读入训练数据
data = pd.read_csv('titanic/train.csv',index_col = 0)
data.head()
```

```python
data.Cabin.value_counts()
```

```python
data.Embarked.value_counts()
```

```python
data.SibSp.value_counts()
```

```python
# 查看数据基本信息
data.info()
```

### 把不需要的三列删除掉

```python
# 删除姓名, 票号, 和货仓号三列
data.drop(columns=['Name','Cabin','Ticket'] , inplace= True)
```

```python
data
```

### 将两个分类变量做数值编码

```python
d = {"male":0  , 'female':1}
data.Sex = data.Sex.map(d)
```

```python
# 生成一个布尔序列, 然后转换成整数型, 将True False 映射成 1, 0 
data.Sex = (data.Sex == 'female').astype('int')
```

```python
# 因为有两个空值, 所有先填充空值, 再转换
# 先填充Embarked这一列
data.Embarked.value_counts()
# S是众数
```

```python
# 采用S填充两个空值
data.Embarked.fillna('S', inplace=True)
```

```python
d = {"S":0 , "C":1 , "Q":2}
data.Embarked = data.Embarked.map(d).astype('int')
```

```python
data.info()
```

```python
# 年龄这一列的空值填充
# 采用年龄的平均值进行 填充
# data.Age.fillna( data.Age.mean() , inplace=True)
```

```python
# data.Age
```

### 分别提取特征数据和标签

```python
# 提取除了Survived列以外的列
X = data.iloc[: , 1:]
```

```python
# 提取Survived列作为标签
y = data.iloc[:, 0]
```

```python
y
```

```python
X
```

### 切割训练集和测试集

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3
,random_state=420)
```

```python
X_train.info()
```

```python
X_test.info()
```

```python
# 分别对训练集和测试集进行填充
# 注意: 我们都采用训练集得到的年龄平均值, 对两个数据进行填充
X_train.Age.fillna(X_train.Age.mean() , inplace= True)
X_test.Age.fillna(X_train.Age.mean() , inplace= True)
```

```python
X_train
```

```python
X_test
```

### 使用决策树模型进行建模

```python
# 实例化模型, 参数为空
clf = DecisionTreeClassifier(random_state=25)
```

```python
# 决策树模型中, 传入的数据中不能包含空值
# 训练模型
clf = clf.fit(X_train, y_train)
```

```python
# 训练集评分
score_train = clf.score(X_train, y_train)
score_train
# 过拟合
```

```python
# 测试集评分
score_test = clf.score(X_test, y_test)
score_test
```

### 学习曲线进行调参

```python
tr = []
te = []
test = []
for i in range(10):
    clf = DecisionTreeClassifier(random_state=25
        ,max_depth=i+1
    )
    clf = clf.fit(X_train, y_train)
    # 模型再训练集上面的准确率
    score_tr = clf.score(X_train,y_train)
    # 模型交叉验证的准确率
    score_te = cross_val_score(clf,X_train,y_train,cv=10).mean()
    tr.append(score_tr)
    te.append(score_te)
    # 模型在测试集上面的准确率
    score_test = clf.score(X_test, y_test)
    test.append(score_test)

print(max(te))
print(te.index(max(te))+1  )
# 训练集分数变化曲线
plt.plot(range(1,11),tr,color="red",label="train")
# 交叉验证分数变化曲线
plt.plot(range(1,11),te,color="blue",label="cross_val_score")
# 测试集分数变化曲线
plt.plot(range(1,11),test,color="green",label="test")
plt.xticks(range(1,11))
plt.legend()
plt.show()
```

### 用网格搜索调整参数GridSearchCV

```python
# 导入网格搜索API
from sklearn.model_selection import GridSearchCV
```

```python
import numpy as np
```

```python
np.linspace(0,0.5 ,20)
```

```python
# 确定参数搜索范围
# max_depth : 1-10
# criterion  基尼和熵两个
#   min_samples_leaf  (1 - 50 , 5)
# min_impurity_decrease  (0 - 0.5  , 走20次)

# 构造网格搜索字典,字典的key用参数名称  , 字典的value 用搜索的序列
d = {"criterion":['gini' , "entropy"] , 
    "max_depth": range(1,11),
    "min_samples_leaf": range(1,51,5), 
    "min_impurity_decrease":np.linspace(0,0.5 ,20)}
```

```python
# 实例化模型
clf = DecisionTreeClassifier(random_state=25)
```

```python
# 网格搜索模型实例化
"""estimator, 评估器: 实例化的算法模型
    param_grid, # 网格搜索参数字典
    scoring=None, # 模型评估的指标, 模型是准确率
    n_jobs=None, # 计算使用的CPU处理器个数
    cv=5 , #  交叉验证次数
    verbose=0, # 调参中的日志输出
    """ 
grid_search = GridSearchCV(clf , param_grid=d , n_jobs=-1, 
             cv=3,  verbose=3   )
```

```python
# 使用网格搜索API进行数据训练
grid_search.fit(X_train , y_train)
```

```python
#属性best_params_查看调整出来的最佳参数
grid_search.best_params_
```

```python
#属性best_score_查看最佳分数
grid_search.best_score_
```

```python
# 网格搜索模型对测试集训练得到的分数
grid_search.score(X_test, y_test)
```
